```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

df=pd.read_csv("Data.csv")
df.head()
```

```python
# Apply OneHotEncoding on the "Country" column
one_hot_encoder = OneHotEncoder()
country_encoded = one_hot_encoder.fit_transform(df[['Country']]).toarray()
country_encoded_df = pd.DataFrame(country_encoded, columns=one_hot_encoder.get_feature_names_out(['Country']))

# Apply LabelEncoding on the "Purchased" column
label_encoder = LabelEncoder()
df['Purchased'] = label_encoder.fit_transform(df['Purchased'])

# Combine the encoded columns with the original dataframe
df = pd.concat([country_encoded_df, df[['Age', 'Salary', 'Purchased']]], axis=1)

print(df)
```

```python
df2=pd.read_csv("Employee.csv")
df2.head()
```

```python
# Apply OneHotEncoding on the "Country" column
one_hot_encoder = OneHotEncoder()
Company_encoded = one_hot_encoder.fit_transform(df2[['Company']]).toarray()
Company_encoded_df2 = pd.DataFrame(Company_encoded, columns=one_hot_encoder.get_feature_names_out(['Company']))
# Apply OneHotEncoding on the "Place" column
one_hot_encoder = OneHotEncoder()
Place_encoded = one_hot_encoder.fit_transform(df2[['Place']]).toarray()
Place_encoded_df2 = pd.DataFrame(Place_encoded, columns=one_hot_encoder.get_feature_names_out(['Place']))

df2['Country'] = df2['Country'].replace('India', 1)

# Combine the encoded columns with the original dataframe
df2 = pd.concat([Company_encoded_df2, df2[['Age', 'Salary']],Place_encoded_df2,df2[['Gender']]], axis=1)

print(df2)
```
